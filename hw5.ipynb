```python
import pandas as pd
import numpy as np

# Load the dataset
data_path = 'Breast_Cancer_dataset.csv'
breast_cancer_data = pd.read_csv(data_path)

# Display the first few rows of the dataset and some basic information about the data
breast_cancer_data
```

```python
stand_data = breast_cancer_data.copy()
```

```python
from sklearn.preprocessing import LabelEncoder


encoder = LabelEncoder()
categorical_cols = stand_data.select_dtypes(include=['object']).columns.drop('Status')
for column in categorical_cols:
    stand_data[column] = encoder.fit_transform(stand_data[column])


stand_data.head()
```

```python
numerical_cols = stand_data.select_dtypes(include=['int64', 'float64']).columns
numerical_cols
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler


# Standardization of numerical columns
scaler = StandardScaler()
stand_data[numerical_cols] = scaler.fit_transform(stand_data[numerical_cols])

stand_data.head()
```

```python
stand_data.describe()
```

```python
stand_data.isnull().sum()
```

```python
# outliers
z_scores = np.abs((stand_data[numerical_cols] - stand_data[numerical_cols].mean()) / stand_data[numerical_cols].std())
threshold = 2
outliers = (z_scores > threshold).any(axis=1)
stand_data = stand_data[~outliers]
```

```python
from sklearn.decomposition import PCA

# Separate the feature columns and the target column
feature = stand_data.drop('Status', axis=1)
target = stand_data['Status']

n_components = min(feature.shape[0], feature.shape[1])

# Initialize PCA, setting the number of components to 5
pca = PCA(n_components=0.95)

# Fit and transform the dataset
principal_components = pca.fit_transform(feature)

# Create a DataFrame for the principal components
pca_df = pd.DataFrame(principal_components, columns=[f'PC{i+1}' for i in range(principal_components.shape[1])])

# Display the first few rows of the PCA transformed data
pca_df['Status'] = target.reset_index(drop=True)
pca_df.head()
```

```python
# Explained variance ratios
explained_variance_ratios = pca.explained_variance_ratio_

# Creating a DataFrame for the ranking of principal components
components = range(1, len(explained_variance_ratios) + 1)
df_pca = pd.DataFrame({
    'Principal Component': components,
    'Explained Variance Ratio': explained_variance_ratios
})

# Calculating cumulative explained variance
df_pca['Cumulative Explained Variance'] = df_pca['Explained Variance Ratio'].cumsum()

# Sorting the DataFrame by explained variance in descending order
df_pca_sorted = df_pca.sort_values(by='Explained Variance Ratio', ascending=False)

# Resetting the index for aesthetics
df_pca_sorted.reset_index(drop=True, inplace=True)

# Displaying the DataFrame
df_pca_sorted
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from imblearn.over_sampling import RandomOverSampler

X = pca_df.drop('Status', axis=1)
y = pca_df['Status']

encoder = LabelEncoder()
y = encoder.fit_transform(y)
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(X, y)

X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

```python
import numpy as np
from scipy.stats import mode

class KNN:
    def __init__(self, k=3):
        self.k = k
        self.X_train = None
        self.y_train = None

    def fit(self, X, y):
        """Fit the training data to the model.

        Parameters:
        X : numpy.ndarray
            Training features.
        y : numpy.ndarray
            Training labels.
        """
        # Ensure the input is a numpy array
        self.X_train = np.array(X)
        self.y_train = np.array(y)

    def predict(self, X):
        """Predict the labels for the input features.

        Parameters:
        X : numpy.ndarray
            Features to predict.

        Returns:
        numpy.ndarray
            Predicted labels.
        """
        # Convert input to numpy array if it's not
        X = np.array(X)

        # Calculate distances from each point in X to all points in X_train
        distances = np.sqrt(((X[:, np.newaxis] - self.X_train) ** 2).sum(axis=2))

        # Get the indices of the k nearest neighbors
        nearest_indices = np.argsort(distances, axis=1)[:, :self.k]

        # Gather the nearest labels
        nearest_labels = self.y_train[nearest_indices]

        # Vote for the most common label
        predictions = mode(nearest_labels, axis=1)[0].flatten()

        return predictions
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.naive_bayes import BernoulliNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier

models = {
    'KNN': KNN(3),
    'Naive Bayes': BernoulliNB(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Gradient Boosting': GradientBoostingClassifier()
}
```

```python
for label, model in models.items():
    model.fit(X_train, y_train)
results = []
for label, model in models.items():
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    results.append([label, accuracy, precision, recall, f1])
evaluate_res = pd.DataFrame(results, columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F1_Score'])
evaluate_res
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score

# Create a scorer for multiple metrics
scoring = {
    'accuracy': make_scorer(accuracy_score),
    'precision': make_scorer(precision_score),
    'recall': make_scorer(recall_score),
    'f1': make_scorer(f1_score)
}

param_grid = {
    'n_estimators': [50, 100, 200],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth': [10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

rf_grid = GridSearchCV(RandomForestClassifier(), param_grid=param_grid, cv=StratifiedKFold(n_splits=5), verbose=2, scoring='accuracy')
rf_grid.fit(X_train, y_train)

print("Best parameters for RF:", rf_grid.best_params_)
print("Best score for RF:", rf_grid.best_score_)
```

```python
gb_params = {
    'n_estimators': [50, 100, 200, 500],
    'learning_rate': [0.01, 0.1, 0.15, 0.2],
    'max_depth': [1,2, 3, 4, 5],
}

gb_grid = GridSearchCV(GradientBoostingClassifier(), gb_params, cv=StratifiedKFold(n_splits=5), scoring=scoring, refit='f1', n_jobs=-1)
gb_grid.fit(X_train, y_train)

print("Best parameters for Gradient Boosting:", gb_grid.best_params_)
print("Best score for Gradient Boosting:", gb_grid.best_score_)
```

```python
# result

final_models = {
    'Random Forest': rf_grid.best_estimator_,
    'Gradient Boosting': gb_grid.best_estimator_
}

final_results = []
for name, model in final_models.items():
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    final_results.append([name, accuracy, precision, recall, f1])

# Display the final results
final_results_df = pd.DataFrame(final_results, columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score'])
print("Tuned Models:")
print(final_results_df)
```
